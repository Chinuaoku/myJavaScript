# FUNDAMENTALS OF JAVASRCIPT
- This is a programming language that is used to make web pages interactive.
- It is an interpreted language: means it doesn't have to be compiled like Java or C or Visual basic and things a lot easier. Javascript code doesnt have to run through a compiler or anything of that nature.
- Runs on the client's computer/browser. Whereas languages like php, ruby-on-rails run on the server. We don't need any kind of server or anything to run it.
- Object based: It is an object oriented program, prototype based instead of class based.

NOTE: Javascript is NOT Java
- They have a similar name but are completely different
- Java is much more complex and a high level language. Typically used to build desktop applications, games etc.
- Same category as C and C++
- Javascript is a "scripting" language meaning that it is interpreted by the web browser when the page is loading.

# HTML vs JAVASCRIPT
- They are completely different and they work together
- HTML is a markup language (display headings, paragraphs, etc) javascript is a programming language
- Javascript uses "events" and "actions" to make your page interactive

# What can you do with Javascript?
- Put content in an HTML page on the fly
- Makw web pages responsive
- Detect a users browser and other information
- Create cookies - holding data in memory
- Validate forms
- Create animations, slideshows, scrollers, etc
- Build apps with Javascript Frameworks (Angular, ReactJS, BackboneJS, etc)
